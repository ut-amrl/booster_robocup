```python
import json
fname = "/home/luisamao/booster_robocup/stat.json"
with open(fname, 'r') as f:
    data = json.load(f)
fname = "/home/luisamao/booster_robocup/examine_joints.json"
with open(fname, 'r') as f:
    joint_data = json.load(f)
```

```python
import matplotlib.pyplot as plt
import numpy as np

actions = np.array(data["actions"])  # shape: (num_steps, num_envs, num_joints)
joints = np.array(joint_data["pos"])    # same shape

num_joints = actions.shape[2]

for i in range(num_joints):
    fig, axs = plt.subplots(1, 2, figsize=(12, 4))

    # Actions
    axs[0].plot(actions[:, 0, i], marker='o', label=f'Action {i}')
    axs[0].set_title(f'Action {i}')
    axs[0].set_xlabel('Step')
    axs[0].set_ylabel('Value')
    axs[0].grid(True)

    # Joint positions
    axs[1].plot(joints[:, 0, i], marker='x', color='r', label=f'Joint {i}')
    axs[1].set_title(f'Joint {i}')
    axs[1].set_xlabel('Step')
    axs[1].set_ylabel('Value')
    axs[1].grid(True)

    plt.tight_layout()
    plt.show()
```
