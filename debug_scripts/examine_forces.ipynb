```python
from utils import *
import json
import numpy as np
```

```python
fname = "../debug_logs/lab_forces.json"
with open(fname, 'r') as f:
    lab_data = json.load(f)
```

```python
lab_data.keys()
body_forces = np.squeeze(np.array(lab_data["body_forces"]))[:,-12:,:]
joint_forces = np.squeeze(np.array(lab_data["joint_forces"]))
print(body_forces.shape, joint_forces.shape)
arr = body_forces.reshape(200, -1)
plot_actions(arr, name = "../debug_logs/lab_body_forces")
plot_actions(joint_forces, name = "../debug_logs/lab_joint_forces")
```
