```python
import json
lab_fname = "../debug_logs/lab_obs_action_floating.json"
gym_fname = "../debug_logs/gym_obs_action_floating.json"
with open(lab_fname, 'r') as f:
    lab_data = json.load(f)
with open(gym_fname, 'r') as f:
    gym_data = json.load(f)
```
